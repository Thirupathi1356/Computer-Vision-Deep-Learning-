```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
!wget -O sample.jpg  https://tse4.mm.bing.net/th/id/OIP.hXWwNOQw15ZVWKlMs-xv0wHaFQ?rs=1&pid=ImgDetMain&o=7&rm=3
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("sample.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
resized = cv2.resize(image, (32, 32))
plt.imshow(resized)
plt.title("Resized image")
plt.axis('off')
```

```python
h, w, _ = image.shape

# Define crop size (for example: 50% of width and height)
crop_h, crop_w = int(h * 0.5), int(w * 0.5)

# Center coordinates
center_y, center_x = h // 2, w // 2

# Calculate start and end points
start_y = center_y - crop_h // 2
end_y = center_y + crop_h // 2
start_x = center_x - crop_w // 2
end_x = center_x + crop_w // 2

# Crop center region
crop = image[start_y:end_y, start_x:end_x]

import matplotlib.pyplot as plt
plt.imshow(crop)
plt.title("Center Cropped Image")
plt.axis('off')
```

```python
edges = cv2.Canny (gray, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')
```
